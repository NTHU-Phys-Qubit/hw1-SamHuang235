# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
n = 1000 # 總粒子數
t = 10000 # 飄移次數
m = 50
x1 = np.zeros([t]) # 追蹤單顆粒子x位置的路徑
y1 = np.zeros([t]) # 追蹤單顆粒子y位置的路徑
x = np.zeros([n]) # 每次Random Walk N次後，x最後的位置
y = np.zeros([n]) # 每次Random Walk N次後，y最後的位置
xy= np.zeros([n]) #距離的平方根
for i in range(n):
    for j in range(t):
        a = np.random.random() - 0.5
        b = np.random.random() - 0.5
        x[i] = x[i] + a
        y[i] = y[i] + b
for i in range(t):
    a = np.random.random() - 0.5
    b = np.random.random() - 0.5
    if i == 0:
        x1[i] = a
        y1[i] = b
    else:
        x1[i] = x1[i-1] + a
        y1[i] = y1[i-1] + b
for i in range(n):
    xy[i] = (x[i]*x[i] + y[i]*y[i])**(1/2)
plt.figure(figsize=(20, 20))
plt.subplot(2,2,1)
plt.plot(x1,y1,'b')
plt.title("2D Random Walk_Method 1")
plt.xlabel('X position')
plt.ylabel('Y position')
plt.subplot(2,2,2)
plt.plot(x,y,'.b')
plt.title("2D Random Walk_Method 1")
plt.xlabel('X position')
plt.ylabel('Y position')
plt.subplot(2,2,3)
plt.hist(xy, bins= m)
plt.title("2D Random Walk_Method 1")
plt.xlabel('Distance')
plt.ylabel('Time')
plt.show()
```

```python
n = 1000 #總粒子數
t = 10000 # 飄移次數
m = 50
x1 = np.zeros([t]) # 追蹤單顆粒子x位置的路徑
y1 = np.zeros([t]) # 追蹤單顆粒子y位置的路徑
x = np.zeros([n]) # 每次Random Walk N次後，x最後的位置
y = np.zeros([n]) # 每次Random Walk N次後，y最後的位置
xy= np.zeros([n]) #距離的平方根
for i in range(n):
    for j in range(t):
        a = np.random.random()
        b = np.random.random()
        if a > 0.5 and b > 0.5:
            x[i] = x[i] + 1
            y[i] = y[i] + 1
        elif a > 0.5 and b <= 0.5:
            x[i] = x[i] + 1
            y[i] = y[i] - 1
        elif a <= 0.5 and b > 0.5:
            x[i] = x[i] - 1
            y[i] = y[i] + 1
        else :
            x[i] = x[i] - 1
            y[i] = y[i] - 1
for i in range(t):
    a = np.random.random()
    b = np.random.random()
    if i == 0:
        if a > 0.5 and b > 0.5:
            x1[i] = 1
            y1[i] = 1
        elif a > 0.5 and b <= 0.5:
            x1[i] = 1
            y1[i] = - 1
        elif a <= 0.5 and b > 0.5:
            x1[i] = - 1
            y1[i] = 1
        else :
            x1[i] = - 1
            y1[i] = - 1
    else:
        if a > 0.5 and b > 0.5:
            x1[i] = x1[i-1] + 0.5
            y1[i] = y1[i-1] + 0.5
        elif a > 0.5 and b <= 0.5:
            x1[i] = x1[i-1] + 0.5
            y1[i] = y1[i-1] - 0.5
        elif a <= 0.5 and b > 0.5:
            x1[i] = x1[i-1] - 0.5
            y1[i] = y1[i-1] + 0.5
        else :
            x1[i] = x1[i-1] - 0.5
            y1[i] = y1[i-1] - 0.5
for i in range(n):
    xy[i] = ( x[i]*x[i] + y[i]*y[i] )**(1/2)
plt.figure(figsize=(20, 20))
plt.subplot(2,2,1)
plt.plot(x1,y1,'b')
plt.title("2D Random Walk_Method 2")
plt.xlabel('X position')
plt.ylabel('Y position')
plt.subplot(2,2,2)
plt.plot(x,y,'.b')
plt.title("2D Random Walk_Method 2")
plt.xlabel('X position')
plt.ylabel('Y position')
plt.subplot(2,2,3)
plt.hist(xy, bins= m)
plt.title("2D Random Walk_Method 2")
plt.xlabel('Distance')
plt.ylabel('Time')
plt.show()
```

```python
n = 1000 # 總粒子數
t = 10000 # 位移次數
m = 50
x1 = np.zeros([t]) # 追蹤單顆粒子x位置的路徑
y1 = np.zeros([t]) # 追蹤單顆粒子y位置的路徑
x = np.zeros([n]) # 每次Random Walk N次後，x最後的位置
y = np.zeros([n]) # 每次Random Walk N次後，y最後的位置
xy= np.zeros([n]) #距離的平方根
for i in range(n): #記錄每顆粒子最後的位置
    for j in range(t):
        a = np.random.random()*2*np.pi
        x[i] = x[i] + np.cos(a)
        y[i] = y[i] + np.sin(a)
for i in range(t): #紀錄單一粒子的軌跡
    a = np.random.random()*2*np.pi
    if i == 0:
        x1[i] = np.cos(a)
        y1[i] = np.sin(a)
    else:
        x1[i] = x1[i-1] + np.cos(a)
        y1[i] = y1[i-1] + np.sin(a)
for i in range(n):
    xy[i] = ( x[i]*x[i] + y[i]*y[i] )**(1/2)
plt.figure(figsize=(20, 20))
plt.subplot(2,2,1)
plt.plot(x1,y1,'b')
plt.title("2D Random Walk_Method 1")
plt.xlabel('X position')
plt.ylabel('Y position')
plt.subplot(2,2,2)
plt.plot(x,y,'.b')
plt.title("2D Random Walk_Method 1")
plt.xlabel('X position')
plt.ylabel('Y position')
plt.subplot(2,2,3)
plt.hist(xy, bins= m)
plt.title("2D Random Walk_Method 1")
plt.xlabel('Distance')
plt.ylabel('Time')
plt.show()
```
