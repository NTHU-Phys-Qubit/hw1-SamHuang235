# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
n = 500
t = 3000
m = 50
p = np.zeros([t]) # 追蹤單顆粒子的路徑
q = np.zeros([n]) # 每次Random Walk N次後，最後的位置
for i in range(n): # 計算每顆粒子最後的位置，並存入q中
    for j in range(t):
        x = np.random.random()
        q[i] = q[i] + x - 0.5   
for i in range(t):
    y = np.random.random() - 0.5
    if i == 0:
        p[i] = y
    else:
        p[i] = p[i-1] + y
plt.figure(figsize=(20, 10))
plt.subplot(1,2,1)
plt.plot(p,'r')
plt.title("1D Random Walk")
plt.xlabel('Orders')
plt.ylabel('Position')
plt.subplot(1,2,2)
plt.hist(q, bins= m)
plt.title("1D Random Walk")
plt.xlabel('Position')
plt.ylabel('Times')
plt.show()
```
